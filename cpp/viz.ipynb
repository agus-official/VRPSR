```python
import numpy as np
import matplotlib.pyplot as plt
from io import BytesIO

h, w = 144, 176
ysz = h * w
csz = (h * w) // 4
psz = (h * w * 3) // 2
with open("akiyo_qcif.decode.yuv", "rb") as f:
    bytes_read = f.read()

yuv = np.frombuffer(bytes_read, dtype=np.uint8).reshape(-1, psz)
plt.imshow(yuv[0][:ysz].reshape(h, w))
plt.show()
print(yuv.shape)
```
